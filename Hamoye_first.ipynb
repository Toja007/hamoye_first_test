```python
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?
df[['Y2017', 'Area']].groupby('Area')['Y2017'].sum().sort_values()
```
